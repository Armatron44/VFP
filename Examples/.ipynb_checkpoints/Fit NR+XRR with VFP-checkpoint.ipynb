### What are we doing?

In this exercise, we'll simulate the reflectivity from an iron coated silicon substrate in D$_{2}$O and H$_{2}$O. <br>
This reflectivity will be simulated as if it were measured on a non-polarised instrument, using the MixedReflectModel component from refnx.reflect. <br>
We will also simulate the XRR from the same substrate measured in air. <br>

This should demonstrate that VFP can handle nuclear, magnetic and imaginary SLDs.

### First we'll define some constants.
We need to fit the density of materials and not the SLDs when fitting across NR and XRR.

```python
# neutron constants.

Si_sl = periodictable.Si.neutron.b_c_complex
SiO2_sl = periodictable.Si.neutron.b_c_complex + 2*periodictable.O.neutron.b_c_complex
Fe_sl = periodictable.Fe.neutron.b_c_complex
FeOx_sl = 2*periodictable.Fe.neutron.b_c_complex + 3*periodictable.O.neutron.b_c_complex

#x-ray constants.
#if sf is function of multiple atoms then we need to seperately define f1 and f2.
Si_sf = periodictable.Si.xray.scattering_factors(wavelength=1.54)
SiO2_sf_f1 = periodictable.Si.xray.scattering_factors(wavelength=1.54)[0] + periodictable.Fe.xray.scattering_factors(wavelength=1.54)[0]
SiO2_sf_f2 = periodictable.Si.xray.scattering_factors(wavelength=1.54)[-1] + periodictable.Fe.xray.scattering_factors(wavelength=1.54)[-1]
Fe_sf = periodictable.Fe.xray.scattering_factors(wavelength=1.54)
FeOx_sf_f1 = 2*periodictable.Fe.xray.scattering_factors(wavelength=1.54)[0]+3*periodictable.O.xray.scattering_factors(wavelength=1.54)[0]
FeOx_sf_f2 = 2*periodictable.Fe.xray.scattering_factors(wavelength=1.54)[-1]+3*periodictable.O.xray.scattering_factors(wavelength=1.54)[-1]

thomp_e = 2.81794E-05 #angstrom
```

```python
#now define some SLDs we will use...

#number density parameters for use
Si_dens = Parameter(0.049930, 'Si_dens', vary=False)
SiO2_dens = Parameter(0.020, 'SiO2_dens', vary=False)
Fe_dens = Parameter(0.0815, 'Fe_dens', vary=False)
FeOx_dens = Parameter(0.017, 'FeOx_dens', vary=False)

#neutron SLDs defined so we can define layers later on.
Si = SLD(2.07, name='Si')
SiO2 = SLD(3.47, name='SiO2')
Fe_Up = SLD(8.02, name='Fe_Up')
Fe_Down = SLD(8.02, name='Fe_Down')
FeOx_Up = SLD(7.0, name='FeOx_Up')
FeOx_Down = SLD(7.0, name='FeOx_Down')
D2O = SLD(6.2, name='D2O')
H2O = SLD(-0.56, name='H2O')

#xray SLDs.
#the precise numbers don't matter here as they'll be re-defined below (bar air).
Si_xray = SLD(complex(thomp_e*Si_sf[0]*Si_dens*1E6, thomp_e*Si_sf[-1]*Si_dens*1E6), name='Si_xray')
SiO2_xray = SLD(3.47, name='SiO2_xray')
Fe_xray = SLD(8.02, name='Fe_xray')
FeOx_xray = SLD(7.0, name='FeOx_xray')
Air = SLD(0.0, name='Air')
```

```python
# these parameters will be used in the NR and XRR contrasts.

Si_SiO2_R = Parameter(3, 'Si_SiO2_rough', vary=False)

SiO2_thick = Parameter(50, 'SiO2_thick', vary=False)
SiO2_Fe_R = Parameter(4, 'SiO2_Fe_R', vary=False)

Fe_thick = Parameter(190, 'Fe_thick', vary=False)
Fe_magmom = Parameter(2.1, 'Fe_magmom', vary=False)
Fe_FeOx_R = Parameter(5, 'FeOx_Fe_rough', vary=False)

FeOx_thick = Parameter(30, 'FeOx_thick', vary=False)
FeOx_m = Parameter(0.5, 'FeOx_m', vary=False)
FeOx_R = Parameter(5, 'FeOx_rough', vary=False)
```

```python
#Neutron layers

Si_lay = Si(0, 0)
Si_lay.sld.real.constraint = 1E-5*Si_sl.real*Si_dens*1E6
Si_lay.sld.imag.constraint = 1E-5*-Si_sl.imag*Si_dens*1E6

SiO2_lay = SiO2(SiO2_thick, Si_SiO2_R)
SiO2_lay.sld.real.constraint = 1E-5*SiO2_sl.real*SiO2_dens*1E6
SiO2_lay.sld.imag.constraint = 1E-5*-SiO2_sl.imag*SiO2_dens*1E6

Fe_up_lay = Fe_Up(Fe_thick, SiO2_Fe_R)
Fe_up_lay.sld.real.constraint = 1E-5*Fe_sl.real*Fe_dens*1E6 + Fe_dens*2.699*1E-5*Fe_magmom*1E6
Fe_up_lay.sld.imag.constraint = 1E-5*-Fe_sl.imag*Fe_dens*1E6

Fe_down_lay = Fe_Down(Fe_thick, SiO2_Fe_R)
Fe_down_lay.sld.real.constraint = 1E-5*Fe_sl.real*Fe_dens*1E6 - Fe_dens*2.699*1E-5*Fe_magmom*1E6
Fe_down_lay.sld.imag.constraint = 1E-5*-Fe_sl.imag*Fe_dens*1E6

FeOx_up_lay = FeOx_Up(FeOx_thick, Fe_FeOx_R)
FeOx_up_lay.sld.real.constraint = 1E-5*FeOx_sl.real*FeOx_dens*1E6 + FeOx_m
FeOx_up_lay.sld.imag.constraint = 1E-5*-FeOx_sl.imag*FeOx_dens*1E6

FeOx_down_lay = FeOx_Down(FeOx_thick, Fe_FeOx_R)
FeOx_down_lay.sld.real.constraint = 1E-5*FeOx_sl.real*FeOx_dens*1E6 - FeOx_m
FeOx_down_lay.sld.imag.constraint = 1E-5*-FeOx_sl.imag*FeOx_dens*1E6

D2O_lay = D2O(0, FeOx_R)

H2O_lay = H2O(0, FeOx_R)

#Xray layers

Si_xray_lay = Si_xray(0, Si_SiO2_R)

SiO2_xray_lay = SiO2_xray(SiO2_thick, SiO2_Fe_R)
SiO2_xray_lay.sld.real.constraint = thomp_e*SiO2_sf_f1*SiO2_dens*1E6
SiO2_xray_lay.sld.imag.constraint = thomp_e*SiO2_sf_f2*SiO2_dens*1E6

Fe_xray_lay = Fe_xray(Fe_thick, Fe_FeOx_R)
Fe_xray_lay.sld.real.constraint = thomp_e*Fe_sf[0]*Fe_dens*1E6
Fe_xray_lay.sld.imag.constraint = thomp_e*Fe_sf[1]*Fe_dens*1E6

FeOx_xray_lay = FeOx_xray(FeOx_thick, FeOx_R)
FeOx_xray_lay.sld.real.constraint = thomp_e*FeOx_sf_f1*FeOx_dens*1E6
FeOx_xray_lay.sld.imag.constraint = thomp_e*FeOx_sf_f2*FeOx_dens*1E6
```

```python
# neutron thickness + roughness

list_of_thickness = (0, SiO2_thick, Fe_thick, FeOx_thick) #first must be zero for Si thickness.
list_of_roughness = (Si_SiO2_R, SiO2_Fe_R, Fe_FeOx_R, FeOx_R)

# xray thickness + roughness

list_of_thickness_xray = (0, FeOx_thick, Fe_thick, SiO2_thick) #first must be zero for Si thickness.
list_of_roughness_xray = (FeOx_R, Fe_FeOx_R, SiO2_Fe_R, Si_SiO2_R)
```

```python
#neutron

"""
The nculear and imaginary SLDs are ordered as the following:
Si, SiO2, Fe, FeOx, D2O, H2O
"""

nuclear_SLDs = np.array([Si_lay.sld.real, SiO2_lay.sld.real, 1E-5*Fe_sl.real*Fe_dens*1E6,
                        1E-5*FeOx_sl.real*FeOx_dens*1E6, D2O_lay.sld.real, H2O_lay.sld.real])
                        
nuclear_iSLDs = np.array([Si_lay.sld.imag, SiO2_lay.sld.imag, 1E-5*-Fe_sl.imag*Fe_dens*1E6, 
                         1E-5*-FeOx_sl.imag*FeOx_dens*1E6, 0, 0]) #just assume the solvent doesnt have imaginary sld.

#magnetic SLD array is length of the number of layers 
#(for example we don't list two magnetic SLDs for H2O & D2O, just one for the solvent layer)
magnetic_SLDs = np.array([0, 0, Fe_dens*2.699*1E-5*Fe_magmom*1E6, FeOx_m, 0])

#create an extent parameter. This is the length of the whole SLD curve so that the erfs approach 1.
VFP_tot = Parameter(name='VFP_tot', constraint=50+(4 * Si_SiO2_R)+SiO2_thick+Fe_thick+FeOx_thick+5+(4 * FeOx_R))
```

```python
#xray

"""
The nculear and imaginary SLDs are ordered as the following:
Air, FeOx, Fe, SiO2, Si
"""

xray_SLDs = np.array([Air.real,
                      FeOx_xray_lay.sld.real,
                      Fe_xray_lay.sld.real,
                      SiO2_xray_lay.sld.real,
                      Si_xray_lay.sld.real])
                        
xray_iSLDs = np.array([0,
                      FeOx_xray_lay.sld.imag,
                      Fe_xray_lay.sld.imag,
                      SiO2_xray_lay.sld.imag,
                      Si_xray_lay.sld.imag])

#magnetic SLD array is length of the number of layers 
#(for example we don't list two magnetic SLDs for H2O & D2O, just one for the solvent layer)
xray_SLDs_mag = np.array([0, 0, 0, 0, 0])

#create an extent parameter. This is the length of the whole SLD curve so that the erfs approach 1.
VFP_tot_xray = Parameter(name='VFP_tot_xray', constraint=50+(4 * FeOx_R)+SiO2_thick+Fe_thick+FeOx_thick+5+(4 * Si_SiO2_R))
```

```python
#create VFPs

D2O_up_vfp = VFP_NR(extent=VFP_tot, 
                    nSLDs=nuclear_SLDs, 
                    niSLDs=nuclear_iSLDs, 
                    mSLDs=magnetic_SLDs, 
                    thicknesses=list_of_thickness,
                    roughnesses=list_of_roughness, 
                    contrast='D2O_up')

D2O_down_vfp = VFP_NR(extent=VFP_tot, 
                      nSLDs=nuclear_SLDs, 
                      niSLDs=nuclear_iSLDs, 
                      mSLDs=magnetic_SLDs, 
                      thicknesses=list_of_thickness,
                      roughnesses=list_of_roughness, 
                      contrast='D2O_down')

H2O_up_vfp = VFP_NR(extent=VFP_tot, 
                    nSLDs=nuclear_SLDs, 
                    niSLDs=nuclear_iSLDs, 
                    mSLDs=magnetic_SLDs, 
                    thicknesses=list_of_thickness,
                    roughnesses=list_of_roughness, 
                    contrast='H2O_up')

H2O_down_vfp = VFP_NR(extent=VFP_tot, 
                      nSLDs=nuclear_SLDs, 
                      niSLDs=nuclear_iSLDs, 
                      mSLDs=magnetic_SLDs, 
                      thicknesses=list_of_thickness,
                      roughnesses=list_of_roughness, 
                      contrast='H2O_down')
```

```python
#xray VFP
Air_XRR_vfp = VFP_XRR(extent=VFP_tot_xray, 
                      nSLDs=xray_SLDs, 
                      niSLDs=xray_iSLDs, 
                      mSLDs=xray_SLDs_mag, 
                      thicknesses=list_of_thickness_xray,
                      roughnesses=list_of_roughness_xray, 
                      contrast='Air')
```

```python
#create stacks
#neutrons
D2O_up_stack = Si_lay | D2O_up_vfp | D2O
D2O_down_stack = Si_lay | D2O_down_vfp | D2O

H2O_up_stack = Si_lay | H2O_up_vfp | H2O
H2O_down_stack = Si_lay | H2O_down_vfp | H2O

#air
Air_stack = Air | Air_XRR_vfp | Si_xray
```

```python
#now define the models

#these are the scale and background parameters
#intensities
D2O_int = Parameter(0.5, 'D2O_int', vary=True, bounds=(0.5*0.8, 0.5*1.2))
H2O_int = Parameter(0.5, 'H2O_int', vary=True, bounds=(0.5*0.8, 0.5*1.2))
#background
D2O_bkg = Parameter(5e-6, 'D2O_bkg', vary=True, bounds=(1e-7, 2e-5))
H2O_bkg = Parameter(8e-6, 'H2O_bkg', vary=True, bounds=(1e-7, 2e-5))

#intensities
Air_int = Parameter(1, 'Air_int', vary=True, bounds=(0.8, 1.2))
#background
Air_bkg = Parameter(1e-7, 'D2O_bkg', vary=True, bounds=(1e-8, 2e-5))

#make sure to pass the normal layer parameters to the auxiliary params of the objectives.
auxiliary_parameters_n = (Si_SiO2_R, SiO2_thick, SiO2_Fe_R, Fe_thick, Fe_magmom, SiO2_dens,
                          Fe_FeOx_R, FeOx_thick, FeOx_m, FeOx_R, Fe_dens, FeOx_dens)
auxiliary_parameters_xray = (Si_SiO2_R, SiO2_thick, SiO2_Fe_R, Fe_thick, SiO2_dens,
                             Fe_FeOx_R, FeOx_thick, FeOx_R, Fe_dens, FeOx_dens)
#define adv_layer interface objective & global objective. These adv layers have their own SLD.
#models
D2O_M = MixedReflectModel((D2O_up_stack, D2O_down_stack), scales=(D2O_int, D2O_int), bkg=D2O_bkg, dq=3*2.355)
H2O_M = MixedReflectModel((H2O_up_stack, H2O_down_stack), scales=(H2O_int, H2O_int), bkg=H2O_bkg, dq=3*2.355)
Air_M = ReflectModel(Air_stack, scale=Air_int, bkg=Air_bkg)
```

```python
#create some noise that is normally distributed & has a standard deviation = 5 % of each reflectivity data point.
#i.e Noise ~ N(0, 0.05*R(Qi))

Q = np.geomspace(9E-3, 3E-1, 100)
D2O_noise = np.random.normal(0, 0.05*D2O_M.model(x=Q), 100)
H2O_noise = np.random.normal(0, 0.05*H2O_M.model(x=Q), 100)
```

```python
Q_xray = np.geomspace(1E-2, 3E-1, 220)
XRR_noise = np.random.normal(0, 0.01*Air_M.model(x=Q_xray), 220) #assume we have 1 % error in our XRR.

#Assume we have used optics with a divergence of 0.01 degrees - FWHM resolution
#note we need dQ and not dQ/Q so we can put this error in Q into the dataset.
angles = np.arcsin((Q_xray*1.54)/(4*np.pi))
resol = np.ones_like(Q_xray)*((0.01*np.pi/180)/(2*np.tan(angles)))*((4*np.pi)/1.54)*np.sin(angles)
```

```python
#add the noise to the simulated reflectivity.
D2O_Randr = D2O_M.model(x=Q)+D2O_noise
H2O_Randr = H2O_M.model(x=Q)+H2O_noise
```

```python
XRR_Randr = Air_M.model(x=Q_xray, x_err=resol)+XRR_noise
```

```python
#check the sld profiles look correct.
#this shows the up structure, where the data points are the calculated SLD values by the VFP.
%matplotlib inline

fig, ax = plt.subplots(2, gridspec_kw=dict(hspace=0.35))
ax[0].plot(D2O_M.structures[0].sld_profile(max_delta_z=0.1)[0]-(50+4*Si_SiO2_R), 
        D2O_M.structures[0].sld_profile(max_delta_z=0.1)[1])

ax[0].plot(D2O_M.structures[0].components[1].get_x_and_y_scatter()[0],
        D2O_M.structures[0].components[1].get_x_and_y_scatter()[1],
        marker='.', linestyle='none')

ax[1].plot(Air_M.structure.sld_profile(max_delta_z=0.1)[0]-(50+4*FeOx_R),
           Air_M.structure.sld_profile(max_delta_z=0.1)[1],
           color='tab:green')

ax[1].plot(Air_M.structure.components[1].get_x_and_y_scatter()[0],
           Air_M.structure.components[1].get_x_and_y_scatter()[1],
           marker='.', linestyle='none', color='tab:red')

ax[0].set_xlabel(r'Distance from Si/SiO$_{2}$ Interface / $\mathrm{\AA{}}$')
ax[0].set_ylabel(r'$\mathrm{SLD}_{\mathrm{n}}+\mathrm{SLD}_{\mathrm{m}}$ / $\mathrm{\AA{}}^{-2} \times 10^{-6}$')
ax[1].set_ylabel(r'$\mathrm{SLD}_{\mathrm{x}}$ / $\mathrm{\AA{}}^{-2} \times 10^{-6}$')
ax[1].set_xlabel(r'Distance from Air/FeO$_{x}$ Interface / $\mathrm{\AA{}}$')
```

```python
# lets see the SLD profile which created the simulated reflectivity & the reflectivity with added noise.

%matplotlib inline

fig, ax = plt.subplots(1, 2, figsize=(12, 5), gridspec_kw=dict(wspace=0.35))
ax[0].plot(D2O_M.structures[0].sld_profile(max_delta_z=0.1)[0]-(50+4*3), 
           (D2O_M.structures[0].sld_profile(max_delta_z=0.1)[1]+D2O_M.structures[1].sld_profile(max_delta_z=0.1)[1])/2,
           label=r'D$_{2}$O SLD$_{\mathrm{n}}$')
ax[0].plot(D2O_M.structures[0].sld_profile(max_delta_z=0.1)[0]-(50+4*3), 
           (D2O_M.structures[0].sld_profile(max_delta_z=0.1)[1]-D2O_M.structures[1].sld_profile(max_delta_z=0.1)[1])/2,
           label=r'SLD$_{\mathrm{m}}$')
ax[0].plot(H2O_M.structures[0].sld_profile(max_delta_z=0.1)[0]-(50+4*3), 
           (H2O_M.structures[0].sld_profile(max_delta_z=0.1)[1]+H2O_M.structures[1].sld_profile(max_delta_z=0.1)[1])/2,
           label=r'H$_{2}$O SLD$_{\mathrm{n}}$')
ax[0].legend(frameon=False)

ax2 = ax[0].twinx()
ax2.plot(Air_M.structure.sld_profile(max_delta_z=0.1)[0]-(5+4*3), #we use the offset used at the end of the SLD profile
         Air_M.structure.sld_profile(max_delta_z=0.1)[1][::-1], #because this is backwards.
         color='tab:red')

ax[0].set_xlabel(r'Distance from Si/SiO2 Interface / $\mathrm{\AA{}}$')
ax[0].set_ylabel(r'$\mathrm{SLD}_{\mathrm{n}}+\mathrm{SLD}_{\mathrm{m}}$ / $\mathrm{\AA{}}^{-2} \times 10^{-6}$')
ax2.set_ylabel(r'$\mathrm{SLD}_{\mathrm{x}}$ / $\mathrm{\AA{}}^{-2} \times 10^{-6}$')
ax2.spines['right'].set_color('tab:red')
ax2.yaxis.label.set_color('tab:red')  
ax2.tick_params(axis='y', colors='tab:red')

ax[1].plot(Q, 0.1*D2O_M.model(x=Q), label='D$_{2}$O', zorder=5)
ax[1].errorbar(x=Q, y=0.1*D2O_Randr, yerr=0.1*np.absolute(D2O_noise), linestyle='none', marker='.', label=r'D$_{2}$O with noise')
ax[1].plot(Q, 0.01*H2O_M.model(x=Q), label='H$_{2}$O', zorder=5)
ax[1].errorbar(x=Q, y=0.01*H2O_Randr, yerr=0.01*np.absolute(H2O_noise), linestyle='none', marker='.', label=r'H$_{2}$O with noise')
ax[1].plot(Q_xray, Air_M.model(x=Q_xray, x_err=resol), label='XRR', zorder=5)
ax[1].errorbar(x=Q_xray, y=XRR_Randr, yerr=np.absolute(XRR_noise), linestyle='none', marker='.', label=r'XRR with noise')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_ylabel(r'$R\left(Q\right)$')
ax[1].set_xlabel(r'$Q$ / $\mathrm{\AA{}}^{-1}$')
ax[1].legend(frameon=False)
```

```python
#now define some SLDs we will use...

#number density parameters for use
Si_dens_fit = Parameter(0.049930, 'Si_dens_fit', vary=False)
SiO2_dens_fit = Parameter(0.020, 'SiO2_dens_fit', vary=True, bounds=(0.015, 0.022))
Fe_dens_fit = Parameter(0.0815, 'Fe_dens_fit', vary=True, bounds=(0.078, 0.085))
FeOx_dens_fit = Parameter(0.017, 'FeOx_dens_fit', vary=True, bounds=(0.015, 0.020))

#neutron SLDs defined so we can define layers later on.
Si_fit = SLD(2.07, name='Si_fit')
SiO2_fit = SLD(3.47, name='SiO2_fit')
Fe_Up_fit = SLD(8.02, name='Fe_Up_fit')
Fe_Down_fit = SLD(8.02, name='Fe_Down_fit')
FeOx_Up_fit = SLD(7.0, name='FeOx_Up_fit')
FeOx_Down_fit = SLD(7.0, name='FeOx_Down_fit')
D2O_fit = SLD(6.2, name='D2O_fit')
H2O_fit = SLD(-0.56, name='H2O_fit')

#xray SLDs.
#the precise numbers don't matter here as they'll be re-defined below (bar air).
Si_xray_fit = SLD(complex(thomp_e*Si_sf[0]*Si_dens_fit*1E6, thomp_e*Si_sf[-1]*Si_dens_fit*1E6), name='Si_xray_fit')
SiO2_xray_fit = SLD(3.47, name='SiO2_xray_fit')
Fe_xray_fit = SLD(8.02, name='Fe_xray_fit')
FeOx_xray_fit = SLD(7.0, name='FeOx_xray_fit')
Air_fit = SLD(0.0, name='Air_fit')
```

```python
# these parameters will be used in the NR and XRR contrasts.

Si_SiO2_R_fit = Parameter(3, 'Si_SiO2_rough_fit', vary=True, bounds=(1, 10))

SiO2_thick_fit = Parameter(50, 'SiO2_thick_fit', vary=True, bounds=(10, 100))
SiO2_Fe_R_fit = Parameter(4, 'SiO2_Fe_R_fit', vary=True, bounds=(1, 10))

Fe_thick_fit = Parameter(190, 'Fe_thick_fit', vary=True, bounds=(150, 250))
Fe_magmom_fit = Parameter(2.1, 'Fe_magmom_fit', vary=True, bounds=(1.7, 2.2))
Fe_FeOx_R_fit = Parameter(5, 'FeOx_Fe_rough_fit', vary=True, bounds=(1, 10))

FeOx_thick_fit = Parameter(30, 'FeOx_thick_fit', vary=True, bounds=(10, 60))
FeOx_m_fit = Parameter(0.5, 'FeOx_m_fit', vary=True, bounds=(0, 1.3))
FeOx_R_fit = Parameter(5, 'FeOx_rough_fit', vary=True, bounds=(1, 10))
```

```python
#Neutron layers to fit

Si_lay_nfit = Si_fit(0, 0)
Si_lay_nfit.sld.real.constraint = 1E-5*Si_sl.real*Si_dens_fit*1E6
Si_lay_nfit.sld.imag.constraint = 1E-5*-Si_sl.imag*Si_dens_fit*1E6

SiO2_lay_nfit = SiO2_fit(SiO2_thick_fit, Si_SiO2_R_fit)
SiO2_lay_nfit.sld.real.constraint = 1E-5*SiO2_sl.real*SiO2_dens_fit*1E6
SiO2_lay_nfit.sld.imag.constraint = 1E-5*-SiO2_sl.imag*SiO2_dens_fit*1E6

Fe_up_lay_nfit = Fe_Up_fit(Fe_thick_fit, SiO2_Fe_R_fit)
Fe_up_lay_nfit.sld.real.constraint = 1E-5*Fe_sl.real*Fe_dens_fit*1E6 + Fe_dens_fit*2.699*1E-5*Fe_magmom_fit*1E6
Fe_up_lay_nfit.sld.imag.constraint = 1E-5*-Fe_sl.imag*Fe_dens_fit*1E6

Fe_down_lay_nfit = Fe_Down_fit(Fe_thick_fit, SiO2_Fe_R_fit)
Fe_down_lay_nfit.sld.real.constraint = 1E-5*Fe_sl.real*Fe_dens_fit*1E6 - Fe_dens_fit*2.699*1E-5*Fe_magmom_fit*1E6
Fe_down_lay_nfit.sld.imag.constraint = 1E-5*-Fe_sl.imag*Fe_dens_fit*1E6

FeOx_up_lay_nfit = FeOx_Up_fit(FeOx_thick_fit, Fe_FeOx_R_fit)
FeOx_up_lay_nfit.sld.real.constraint = 1E-5*FeOx_sl.real*FeOx_dens*1E6 + FeOx_m_fit
FeOx_up_lay_nfit.sld.imag.constraint = 1E-5*-FeOx_sl.imag*FeOx_dens*1E6

FeOx_down_lay_nfit = FeOx_Down_fit(FeOx_thick_fit, Fe_FeOx_R_fit)
FeOx_down_lay_nfit.sld.real.constraint = 1E-5*FeOx_sl.real*FeOx_dens_fit*1E6 - FeOx_m_fit
FeOx_down_lay_nfit.sld.imag.constraint = 1E-5*-FeOx_sl.imag*FeOx_dens_fit*1E6

D2O_lay_nfit = D2O_fit(0, FeOx_R_fit)
D2O_lay_nfit.sld.real.setp(vary=True, bounds=(6, 6.4))

H2O_lay_nfit = H2O_fit(0, FeOx_R_fit)

#Xray layers

Si_xray_lay_fit = Si_xray_fit(0, Si_SiO2_R_fit)

SiO2_xray_lay_fit = SiO2_xray_fit(SiO2_thick_fit, SiO2_Fe_R_fit)
SiO2_xray_lay_fit.sld.real.constraint = thomp_e*SiO2_sf_f1*SiO2_dens_fit*1E6
SiO2_xray_lay_fit.sld.imag.constraint = thomp_e*SiO2_sf_f2*SiO2_dens_fit*1E6

Fe_xray_lay_fit = Fe_xray_fit(Fe_thick_fit, Fe_FeOx_R_fit)
Fe_xray_lay_fit.sld.real.constraint = thomp_e*Fe_sf[0]*Fe_dens_fit*1E6
Fe_xray_lay_fit.sld.imag.constraint = thomp_e*Fe_sf[1]*Fe_dens_fit*1E6

FeOx_xray_lay_fit = FeOx_xray(FeOx_thick_fit, FeOx_R_fit)
FeOx_xray_lay_fit.sld.real.constraint = thomp_e*FeOx_sf_f1*FeOx_dens_fit*1E6
FeOx_xray_lay_fit.sld.imag.constraint = thomp_e*FeOx_sf_f2*FeOx_dens_fit*1E6
```

```python
# neutron thickness + roughness

list_of_thickness_nfit = (0, SiO2_thick_fit, Fe_thick_fit, FeOx_thick_fit) #first must be zero for Si thickness.
list_of_roughness_nfit = (Si_SiO2_R_fit, SiO2_Fe_R_fit, Fe_FeOx_R_fit, FeOx_R_fit)

# xray thickness + roughness

list_of_thickness_xray_fit = (0, FeOx_thick_fit, Fe_thick_fit, SiO2_thick_fit) #first must be zero for Air thickness.
list_of_roughness_xray_fit = (FeOx_R_fit, Fe_FeOx_R_fit, SiO2_Fe_R_fit, Si_SiO2_R_fit)
```

```python
#neutron fit

"""
The nuclear and imaginary SLDs are ordered as the following:
Si, SiO2, Fe, FeOx, D2O, H2O
"""

nuclear_SLDs_fit = np.array([Si_lay_nfit.sld.real, SiO2_lay_nfit.sld.real, 1E-5*Fe_sl.real*Fe_dens_fit*1E6,
                        1E-5*FeOx_sl.real*FeOx_dens_fit*1E6, D2O_lay_nfit.sld.real, H2O_lay_nfit.sld.real])
                        
nuclear_iSLDs_fit = np.array([Si_lay_nfit.sld.imag, SiO2_lay_nfit.sld.imag, 1E-5*-Fe_sl.imag*Fe_dens_fit*1E6, 
                         1E-5*-FeOx_sl.imag*FeOx_dens_fit*1E6, 0, 0]) #just assume the solvent doesnt have imaginary sld.

#magnetic SLD array is length of the number of layers 
#(for example we don't list two magnetic SLDs for H2O & D2O, just one for the solvent layer)
magnetic_SLDs_fit = np.array([0, 0, Fe_dens_fit*2.699*1E-5*Fe_magmom_fit*1E6, FeOx_m_fit, 0])

#create an extent parameter. This is the length of the whole SLD curve so that the erfs approach 1.
VFP_tot_fit = Parameter(name='VFP_tot_fit', constraint=50+(4 * Si_SiO2_R_fit)+SiO2_thick_fit+Fe_thick_fit+FeOx_thick_fit+5+(4 * FeOx_R_fit))
```

```python
#xray fit

"""
The nuclear and imaginary SLDs are ordered as the following:
Air, FeOx, Fe, SiO2, Si
"""

xray_SLDs_fit = np.array([Air_fit.real,
                          FeOx_xray_lay_fit.sld.real,
                          Fe_xray_lay_fit.sld.real,
                          SiO2_xray_lay_fit.sld.real,
                          Si_xray_lay_fit.sld.real])
                        
xray_iSLDs_fit = np.array([0,
                           FeOx_xray_lay_fit.sld.imag,
                           Fe_xray_lay_fit.sld.imag,
                           SiO2_xray_lay_fit.sld.imag,
                           Si_xray_lay_fit.sld.imag])

#magnetic SLD array is length of the number of layers 
#(for example we don't list two magnetic SLDs for H2O & D2O, just one for the solvent layer)
xray_SLDs_mag_fit = np.array([0, 0, 0, 0, 0])

#create an extent parameter. This is the length of the whole SLD curve so that the erfs approach 1.
VFP_tot_xray_fit = Parameter(name='VFP_tot_xray_fit', constraint=50+(4 * FeOx_R_fit)+SiO2_thick_fit+Fe_thick_fit+FeOx_thick_fit+5+(4 * Si_SiO2_R_fit))
```

```python
#create VFPs

D2O_up_vfp_fit = VFP_NR(extent=VFP_tot_fit, 
                        nSLDs=nuclear_SLDs_fit, 
                        niSLDs=nuclear_iSLDs_fit, 
                        mSLDs=magnetic_SLDs_fit, 
                        thicknesses=list_of_thickness_nfit,
                        roughnesses=list_of_roughness_nfit, 
                        contrast='D2O_up')

D2O_down_vfp_fit = VFP_NR(extent=VFP_tot_fit, 
                          nSLDs=nuclear_SLDs_fit, 
                          niSLDs=nuclear_iSLDs_fit, 
                          mSLDs=magnetic_SLDs_fit, 
                          thicknesses=list_of_thickness_nfit,
                          roughnesses=list_of_roughness_nfit, 
                          contrast='D2O_down')

H2O_up_vfp_fit = VFP_NR(extent=VFP_tot_fit, 
                        nSLDs=nuclear_SLDs_fit, 
                        niSLDs=nuclear_iSLDs_fit, 
                        mSLDs=magnetic_SLDs_fit, 
                        thicknesses=list_of_thickness_nfit,
                        roughnesses=list_of_roughness_nfit, 
                        contrast='H2O_up')

H2O_down_vfp_fit = VFP_NR(extent=VFP_tot_fit, 
                          nSLDs=nuclear_SLDs_fit, 
                          niSLDs=nuclear_iSLDs_fit, 
                          mSLDs=magnetic_SLDs_fit, 
                          thicknesses=list_of_thickness_nfit,
                          roughnesses=list_of_roughness_nfit, 
                          contrast='H2O_down')
```

```python
#xray VFP
Air_XRR_vfp_fit = VFP_XRR(extent=VFP_tot_xray_fit, 
                          nSLDs=xray_SLDs_fit, 
                          niSLDs=xray_iSLDs_fit, 
                          mSLDs=xray_SLDs_mag_fit, 
                          thicknesses=list_of_thickness_xray_fit,
                          roughnesses=list_of_roughness_xray_fit, 
                          contrast='Air')
```

```python
#create stacks
#neutrons
D2O_up_stack_fit = Si_lay_nfit | D2O_up_vfp_fit | D2O_fit
D2O_down_stack_fit = Si_lay_nfit | D2O_down_vfp_fit | D2O_fit

H2O_up_stack_fit = Si_lay_nfit | H2O_up_vfp_fit | H2O_fit
H2O_down_stack_fit = Si_lay_nfit | H2O_down_vfp_fit | H2O_fit

#air
Air_stack_fit = Air_fit | Air_XRR_vfp_fit | Si_xray_fit
```

```python
#now define the models

#these are the scale and background parameters
#intensities
D2O_int_fit = Parameter(0.5, 'D2O_int_fit', vary=True, bounds=(0.5*0.8, 0.5*1.2))
H2O_int_fit = Parameter(0.5, 'H2O_int_fit', vary=True, bounds=(0.5*0.8, 0.5*1.2))
#background
D2O_bkg_fit = Parameter(5e-6, 'D2O_bkg_fit', vary=True, bounds=(1e-7, 2e-5))
H2O_bkg_fit = Parameter(8e-6, 'H2O_bkg_fit', vary=True, bounds=(1e-7, 2e-5))

#intensities
Air_int_fit = Parameter(1, 'Air_int_fit', vary=True, bounds=(0.8, 1.2))
#background
Air_bkg_fit = Parameter(1e-7, 'D2O_bkg_fit', vary=True, bounds=(1e-8, 2e-5))

#make sure to pass the normal layer parameters to the auxiliary params of the objectives.
auxiliary_parameters_n_fit = (Si_SiO2_R_fit, SiO2_thick_fit, SiO2_Fe_R_fit, Fe_thick_fit, Fe_magmom_fit, SiO2_dens_fit,
                              Fe_FeOx_R_fit, FeOx_thick_fit, FeOx_m_fit, FeOx_R_fit, Fe_dens_fit, FeOx_dens_fit)
auxiliary_parameters_xray_fit = (Si_SiO2_R_fit, SiO2_thick_fit, SiO2_Fe_R_fit, Fe_thick_fit, SiO2_dens_fit,
                                 Fe_FeOx_R_fit, FeOx_thick_fit, FeOx_R_fit, Fe_dens_fit, FeOx_dens_fit)
#define adv_layer interface objective & global objective. These adv layers have their own SLD.
#models
D2O_M_fit = MixedReflectModel((D2O_up_stack_fit, D2O_down_stack_fit), scales=(D2O_int_fit, D2O_int_fit), bkg=D2O_bkg_fit, dq=3*2.355)
H2O_M_fit = MixedReflectModel((H2O_up_stack_fit, H2O_down_stack_fit), scales=(H2O_int_fit, H2O_int_fit), bkg=H2O_bkg_fit, dq=3*2.355)
Air_M_fit = ReflectModel(Air_stack_fit, scale=Air_int_fit, bkg=Air_bkg_fit, dq_type='pointwise')
```

```python
#remember to pass the auxiliary_parameters tuple to the objective!

D2O_obj_fit = Objective(D2O_M_fit, D2O_Data_rfnx, auxiliary_params=auxiliary_parameters_n_fit)
H2O_obj_fit = Objective(H2O_M_fit, H2O_Data_rfnx, auxiliary_params=auxiliary_parameters_n_fit)
XRR_obj_fit = Objective(Air_M_fit, XRR_Data_rfnx, auxiliary_params=auxiliary_parameters_xray_fit)

glob_obj = GlobalObjective([D2O_obj_fit, H2O_obj_fit, XRR_obj_fit])
```

```python
#lets take a look at the fit.

from dynesty import utils as dyfunc
from functools import reduce

samples = res_load.samples  # samples explored by nested sampling.
weights = np.exp(res_load.logwt - res_load.logz[-1])  # normalized weights. These glob_obj from low to high.

# Compute 2.5%-97.5% quantiles.
quantiles = [dyfunc.quantile(samps, [0.025, 0.5, 0.975], weights=weights)
             for samps in samples.T]

#lets print out the posterior found by NS.
for i, j in enumerate(quantiles):
    print(f"""{labelz[i]}
    Median: {j[1]:.4E}
    Lower: {j[0]:.4E}
    Upper: {j[2]:.4E}
    Minus: {j[1]-j[0]:.4E}
    Plus: {j[2]-j[1]:.4E}""")

#function to return the posterior of the samples.
def get_posterior(quantiles, samples, sample_len=300):
    """
    This function finds the samples from the nested sampling that lie within the 95 % CI for all parameters.
    First, the indexes of the parameter values between the individual 95 % CI are found.
    Then the indexes where all parameter values fall within the overall posterior.
    The full posterior sample is then returned. Finally, we take a sub-sample of the posterior.
    """
    indiv_index = []
    for i, j in enumerate(samples.T): #samples.T is (i x n) where i is the number of parameters & n is number of samples.
        indexes = np.where((j >= quantiles[i][0]) & (j <= quantiles[i][2]))[0] # return just the array.
        indiv_index.append(indexes) # a list of indexes of different sizes.
    
    #now find those samples that all within the posterior.
    intersect_tuple = tuple(i for i in indiv_index) # make tuple out of list of indexes.
    posterior_indexes = reduce(np.intersect1d, intersect_tuple) # now we have array of indicies of samples that are all within posterior.
    full_posterior = samples[posterior_indexes] # full posterior.
    
    if sample_len > len(posterior_indexes):
        sample_len = len(posterior_indexes)
        
    random_number = np.random.choice(posterior_indexes, size=sample_len, replace=False)
    sampled_posterior = samples[random_number]
    return sampled_posterior

#return 300 of the posterior samples.
posterior_samples = get_posterior(quantiles, samples, 300)

#set the varying_parameters stored in glob_obj to the median values found in quantiles.
for i, j in enumerate(labelz):
    if j == GO.varying_parameters()[i].name:
        GO.varying_parameters()[i].value = quantiles[i][1]
```
